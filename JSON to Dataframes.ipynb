```python
import pandas as pd
from pandas.io.json import json_normalize
import json
from datetime import datetime
import numpy as np
```

```python
# Load in json data
receipts = pd.read_json("receipts.json", lines = True)
brands = pd.read_json("brands.json", lines = True)
users = pd.read_json("users.json", lines = True)
```

## Receipts

```python
receipts
```

```python
receipts["_id"] = receipts["_id"].astype(str).str[10:-2]

# make createDate into a date
receipts["createDate"] = receipts["createDate"].astype(str).str[10:-1]
receipts["createDate"] = receipts["createDate"].astype(float)/1000
receipts["createDate"] = receipts["createDate"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# make dateScanned into a date
receipts["dateScanned"] = receipts["dateScanned"].astype(str).str[10:-1]
receipts["dateScanned"] = receipts["dateScanned"].astype(float)/1000
receipts["dateScanned"] = receipts["dateScanned"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# make finishedDate into a date
receipts["finishedDate"] = receipts["finishedDate"].astype(str).str[10:-1]
# some values are missing, so replace those with NaN 
receipts.finishedDate.replace("", np.nan, inplace = True)
# replace the NaN values to 0000000000000 so that a date will be created
receipts.finishedDate.fillna("0000000000000", inplace = True)
receipts["finishedDate"] = receipts["finishedDate"].astype(float)/1000
receipts["finishedDate"] = receipts["finishedDate"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# make finishedDate into a date
receipts["modifyDate"] = receipts["modifyDate"].astype(str).str[10:-1]
receipts["modifyDate"] = receipts["modifyDate"].astype(float)/1000
receipts["modifyDate"] = receipts["modifyDate"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# make finishedDate into a date
receipts["pointsAwardedDate"] = receipts["pointsAwardedDate"].astype(str).str[10:-1]
receipts.pointsAwardedDate.replace("", np.nan, inplace = True)
receipts.pointsAwardedDate.fillna("0000000000000", inplace = True)
receipts["pointsAwardedDate"] = receipts["pointsAwardedDate"].astype(float)/1000
receipts["pointsAwardedDate"] = receipts["pointsAwardedDate"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# make finishedDate into a date
receipts["purchaseDate"] = receipts["purchaseDate"].astype(str).str[10:-1]
# similar to before, replace any blank values to NaN and change them to 0000000000000
receipts.purchaseDate.replace("", np.nan, inplace = True)
receipts.purchaseDate.fillna("0000000000000", inplace = True)
receipts["purchaseDate"] = receipts["purchaseDate"].astype(float)/1000
receipts["purchaseDate"] = receipts["purchaseDate"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# Replace any values that were initially missing to NaT
receipts.loc[receipts["finishedDate"] == "01-01-1970", "finishedDate"] = pd.NaT
receipts.loc[receipts["pointsAwardedDate"] == "01-01-1970", "pointsAwardedDate"] = pd.NaT
receipts.loc[receipts["purchaseDate"] == "01-01-1970", "purchaseDate"] = pd.NaT

receipts
```

```python
# Save to a csv to use later for data exploration
receipts.to_csv("receipts.csv", index = False)
```

## Brands

```python
brands["_id"] = brands["_id"].astype(str).str[10:-2]

# the cpg column is in a dictionary. This takes it out of that format so that we can use the cpg information
temp = brands["cpg"].apply(pd.Series)
brands["cpg_id"] = temp["$id"]
brands["cpg_ref"] = temp["$ref"]
brands["cpg_id"] = brands["cpg_id"].astype(str).str[10:-2]
brands
```

```python
brands.to_csv("brands.csv", index = False)
```

```python
brands.fillna('', inplace = True)
brands["cpg"] = brands["cpg"].astype(str)
brands["barcode"] = brands["barcode"].astype(str)
brands["topBrand"] = brands["topBrand"].astype(str)
brands
```

## Users

```python
# remove the unnecessary text from each of the columns
users["_id"] = users["_id"].astype(str).str[10:-2]
users["createdDate"] = users["createdDate"].astype(str).str[10:-1]
users["lastLogin"] = users["lastLogin"].astype(str).str[10:-1]

# make the createdDate into a date
users["createdDate"] = users["createdDate"].astype(float)/1000
users["createdDate"] = users["createdDate"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# make the lastLogin into a date
# similar to what was done in the brands, replace blank values with NaN and then switch with 0000000000000
users.lastLogin.replace("", np.nan, inplace = True)
users.lastLogin.fillna("0000000000000", inplace = True)
users["lastLogin"] = users["lastLogin"].astype(float)/1000
users["lastLogin"] = users["lastLogin"].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))

# changes any values that were originally missing to be NaT
users.loc[users["lastLogin"] == "01-01-1970", "lastLogin"] = pd.NaT

users
```

```python
# Save to a csv to use later for data exploration
users.to_csv("users.csv", index = False)
```
